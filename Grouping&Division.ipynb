Grouping (Division and Distribution)

Number of ways of dividing n-different objects into r groups of group size n1 n2 n3 ……nr, where no two groups have same size. (i.e, ni != nj, for any i,j), will be

 n!/ (n1! * n2! * n3!.....nr!) 

eg, Dividing 10 different objects into groups of 5,3 and 2 = 10! /(5!.3!.2!) (Note : these groups have to be different sizes)

If n different objects have  to be divided into groups in which k1 have group size n1 and k2 number of groups have size n2, kr groups have group size nr, then number of ways will be 

n! / [(n1!)k1 . ( n2)k2…….(nr)kr (k1).k2)…..(kr)]

where n = k1n1 + k2n2 + ….+ krnr

8 have to be divided into 3,3 and 2 then: the ways will be – 8!/ 3!.3!.2! 2!

Distribution

1.If n different objects are to are to be distributed among n, one object to each; number of ways = n!

2.If n different objects are to be distributed among r persons, no restriction but all objects should be distributed, same person can get one or more objects: rn

Circular Permutation:

If n different objects are to be permuted in a circle, then number of ways:
a)	(n-1)! If clockwise and anticlockwise can be distinguished
b)	(n-1)! /2 , if clockwise and anticlockwise cannot be distinguished
